##### **1. Figure 6 - Main manuscript**

- a. World diesel market regional share [%]

```python
#Read data
shares_breakdown = pd.read_excel('Diesel_regional_shares.xlsx', index_col=0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import StrMethodFormatter

# Set plot style
sns.set_style("whitegrid")
sns.set_style("ticks")

# Get unique values of Pathway and market
pathways = shares_breakdown["Database"].unique()

# Create the subplots
fig, axs = plt.subplots(ncols=len(pathways), figsize=(12, 3.8), sharex=True, sharey=True)
# Define custom colors
colors = ["#3f7489ff", "#f3964bff", "#90ee90ff", "#c35544ff","#a39fa8ff","#6129a8ff","#333333ff","#299c29ff","#c39993ff","#5caaca80","#0aadee80", "#1a6d14e3"]

dict_path = {'SSP2-Base': '3.5 C', 'SSP2-PkBudg1150': '2 C', 'SSP2-PkBudg500': '1.5 C'}
# Loop through each pathway and market combination
for i, pathway in enumerate(pathways):
    for keys in dict_path:
        if pathway == keys:
            path_C = dict_path.get(keys)
    # Filter the data for the current pathway 
    subset = shares_breakdown[(shares_breakdown["Database"] == pathway)]

    # Create the stacked area plot
    axs[i].stackplot(subset["Year"], subset["JPN"], subset["OAS"], subset["CAZ"], subset["EUR"], subset["SSA"], subset["MEA"], subset["IND"], 
                            subset["USA"], subset["REF"],  subset["NEU"], subset["LAM"], subset["CHA"], labels=["JPN", "OAS", "CAZ", "EUR", "SSA", 
                                                                                                               "MEA", "IND", "USA", "REF","NEU", 
                                                                                                               "LAM", "CHA"], colors=colors, alpha=0.5)

    # Set the title and axis labels
    if i == 0:
        axs[i].set_ylabel(f"World diesel \nmarket regional share [%]", fontsize=14, fontweight="bold", fontname="Arial")
    axs[i].tick_params(axis='y', which='major', labelsize=12, width=2, length=8)
    axs[i].tick_params(axis='x', which='major', labelsize=12, width=2, length=8, rotation=90)
    axs[i].set_title(f"{path_C}", fontsize=14, fontweight="bold", fontname="Arial")

    for spine in axs[i].spines.values():
        spine.set_linewidth(2)
            
# Add a legend to the last subplot
plt.legend(fontsize=12, loc='upper right', bbox_to_anchor=(1.9, 1), ncol=2, frameon=False)

# Adjust the spacing between subplots
plt.tight_layout()
plt.subplots_adjust(wspace=0.2)  

plt.show()
```

##### **2. Figure 7 - Main manuscript**

```python
results_dynamic = pd.read_excel('Dynamic_profiles.xlsx', index_col=0)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set default font to Arial
mpl.rcParams['font.family'] = 'Arial'

# Filter data for each Database value
results_base = results_dynamic[results_dynamic['Database'] == 'SSP2-Base']
results_pkbudg500 = results_dynamic[results_dynamic['Database'] == 'SSP2-PkBudg500']

# Define colors and line styles
color_map = {
    'fossil': '#07040dff',
    'DAC/BTH': '#683eb4ff',
    'DAC/Wind': '#f3964bff'
}
line_style_map = {
    'Non-integrated': 'solid',
    'Integrated': 'dashed'
}

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(6, 4.5))

# Function to plot data
def plot_data(ax, data, temp):
    for technology in data['Technology '].unique():
        for type_ in data['Type'].unique():
            subset = data[(data['Technology '] == technology) & (data['Type'] == type_)]
            if not subset.empty:
                ax.plot(subset['Year'], subset['Total'], 
                        label=f'{technology} ({type_})', 
                        color=color_map[technology], 
                        linestyle=line_style_map[type_],
                        linewidth=3)  # Make line width thicker
    ax.set_title(f'{temp}$^{{\\mathbf{{\circ}}}}$C', fontweight="bold", fontsize=14)
    ax.set_xlabel('')  # Remove x-axis label
    ax.grid(True, linewidth=0.3)  # Add grid lines with custom linewidth

    # Customize the frame (axes border) linewidth
    for spine in ax.spines.values():
        spine.set_linewidth(1.5)

    # Customize the tick width
    ax.tick_params(axis='both', width=1.5)

# Plot data for SSP2-Base
plot_data(axes[0], results_base, '3.5')
axes[0].set_ylabel('Global warming\n[kg CO$\mathbf{_{2-eq.}}$ tkm$\mathbf{^{-1}}$]', fontweight='bold', fontsize=14)

# Plot data for SSP2-PkBudg500
plot_data(axes[1], results_pkbudg500, '1.5')
axes[1].set_ylabel('')  # Remove ylabel
axes[1].tick_params(axis='y', which='both', labelleft=False)  # Remove y-axis labels but keep ticks

# Harmonize y-axis limits
y_min = min(axes[0].get_ylim()[0], axes[1].get_ylim()[0])
y_max = max(axes[0].get_ylim()[1], axes[1].get_ylim()[1])
axes[0].set_ylim(y_min, y_max)
axes[1].set_ylim(y_min, y_max)

# Collect legend handles and labels
handles, labels = [], []
for ax in axes:
    for handle, label in zip(*ax.get_legend_handles_labels()):
        if label not in labels:
            handles.append(handle)
            labels.append(label)

# Create a single legend at the bottom without a frame and with Arial font
fig.legend(handles, labels, loc='lower center', ncol=2, fontsize=14, frameon=False, prop={'family': 'Arial'})

plt.tight_layout(rect=[0.02, 0.15, 1, 1])  # Adjust layout to make space for the legend
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
from matplotlib.ticker import ScalarFormatter

# Create the 3 dataframes
results_break_500 = results[results['Database'].isin(['SSP2-PkBudg500'])]
results_break_500_2020 = results_break_500[results_break_500['Year'].isin([2020])]
results_break_500_2030 = results_break_500[results_break_500['Year'].isin([2030])]
results_break_500_2050 = results_break_500[results_break_500['Year'].isin([2050])]

min_limit = -0.2  # Replace with your desired minimum limit
max_limit = 0.15  # Replace with your desired maximum limit

color = ['#1dabcdff', '#8763c9ff', '#f37e27ff', '#f37e27ff', '#2b4e1eff', '#979191c8']
fig, axes = plt.subplots(1,3, figsize=(15,4))

#Plotting the 2020 results
ax = results_break_500_2020.iloc[:,1:7].plot(kind="bar", stacked=True, width = 0.5, color = color, ec = "black", linewidth = 0.4,ax=axes[0], legend=False)
df2020 = results_break_500_2020.iloc[:,1:7]
df2020 = df2020.transpose()
df2020.columns = ["Non-integrated","Integrated","Non-integrated","Integrated"]
df2020.sum(axis=0).plot(ax=ax,marker='D',color='black',linestyle='None', markersize = 5, label='Total impact')
ax.set_xticklabels(["Non-integrated","Integrated","Non-integrated","Integrated"], fontsize=12, font='Arial', rotation=0)
ax.axhline(0, color='black', linewidth=2)
ax.axvline(1.5, color='black', linewidth=2)
ax.set_ylim(min_limit, max_limit)
ax.set_ylabel('1.5$\mathbf{^{o}}$C\nGlobal warming\n[kg CO$\mathbf{_{2-eq.}}$ tkm$\mathbf{^{-1}}$]', fontweight="bold",fontsize=14, fontname="Arial")
ax.set_title('2020', fontweight="bold", fontsize=14, fontname="Arial")

#Plotting the 2030 results
ax1 = results_break_500_2030.iloc[:,1:7].plot(kind="bar", stacked=True, width = 0.5, color = color, ec = "black", linewidth = 0.4,ax=axes[1], legend=False)
df2030 = results_break_500_2030.iloc[:,1:7]
df2030 = df2030.transpose()
df2030.columns = ["Non-integrated","Integrated","Non-integrated","Integrated"]
df2030.sum(axis=0).plot(ax=ax1,marker='D',color='black',linestyle='None',markersize = 5, label='Total impact')
ax1.set_xticklabels(["Non-integrated","Integrated","Non-integrated","Integrated"], fontsize=12, font='Arial', rotation=0)
ax1.axhline(0, color='black', linewidth=2)
ax1.axvline(1.5, color='black', linewidth=2)
ax1.set_ylim(min_limit, max_limit)
# Plotting the sum of 'Integrated' for 2020
df2020.loc[:, df2020.columns != 'Integrated'] = 0
df2020.sum(axis=0).plot(ax=ax1,marker='*',color='#3d38d8ff',linestyle='None', markersize = 10, label='integrated LCA impact in 2020')
ax1.set_title('2030', fontweight="bold", fontsize=14, fontname="Arial")

#Plotting the 2030 results
ax2 = results_break_500_2050.iloc[:,1:7].plot(kind="bar", stacked=True, width = 0.5, color = color, ec = "black", linewidth = 0.4,ax=axes[2], legend=False)
df2050 = results_break_500_2050.iloc[:,1:7]
df2050 = df2050.transpose()
df2050.columns = ["Non-integrated","Integrated","Non-integrated","Integrated"]
df2050.sum(axis=0).plot(ax=ax2,marker='D',color='black',linestyle='None', markersize = 5, label='Total impact')
ax2.set_xticklabels(["Non-integrated","Integrated","Non-integrated","Integrated"], fontsize=12, font='Arial', rotation=0)
ax2.axhline(0, color='black', linewidth=2)
ax2.axvline(1.5, color='black', linewidth=2)
ax2.set_ylim(min_limit, max_limit)
# Plotting the sum of 'Integrated' for 2020
df2020.sum(axis=0).plot(ax=ax2,marker='*',color='#3d38d8ff',linestyle='None', markersize = 10, label='Integrated LCA impact in 2020')
ax2.set_title('2050', fontweight="bold", fontsize=14, fontname="Arial")

legend = plt.legend(loc = 'lower center', bbox_to_anchor = (-0.8, -0.3), frameon = False, ncol = 5)

#Manually add the text
fig.text(0.185, 0.82, 'DAC/Wind', ha='center', fontsize=14, fontweight="bold", fontstyle='italic', font='Arial')
fig.text(0.298, 0.82, 'DAC/BTH', ha='center', fontsize=14, fontweight="bold", fontstyle='italic', font='Arial')
fig.text(0.455, 0.82, 'DAC/Wind', ha='center', fontsize=14, fontweight="bold", fontstyle='italic', font='Arial')
fig.text(0.57, 0.82, 'DAC/BTH', ha='center', fontsize=14, fontweight="bold", fontstyle='italic', font='Arial')
fig.text(0.73, 0.82, 'DAC/Wind', ha='center', fontsize=14, fontweight="bold", fontstyle='italic', font='Arial')
fig.text(0.845, 0.82, 'DAC/BTH', ha='center', fontsize=14, fontweight="bold", fontstyle='italic', font='Arial')

plt.show()
```
